```python
import numpy
import pandas as pd
import math

def gradientdescent(x,y):
    mcurr,bcurr=0,0
    iterations =200000
    rate = 0.0004
    m = len(x)
    prev_cost = 0

    for i in range(iterations):
        yp = mcurr*x+bcurr
        cost = (1/(2*m))*sum([j**2 for j in (y-yp)])
        pd_mcurr = -(1/m)*sum(x*(y-yp))
        pd_bcurr = -(1/m)*sum(y-yp)        
        mcurr=mcurr-rate*pd_mcurr
        bcurr=bcurr-rate*pd_bcurr
        if(i%1000==0):
            print("m:{} b:{} cost:{} iteration:{}".format(mcurr,bcurr,cost,i))
        if(math.isclose(prev_cost,cost,rel_tol=1e-09)):
            print("\n","The required m and b are {} and {}".format(mcurr,bcurr))
            break;
        prev_cost = cost

    return mcurr,bcurr

df = pd.read_csv("/kaggle/input/student-scores/test_scores.csv")
x=numpy.array(df.math)
y = numpy.array(df.cs)

weight,bias = gradientdescent(x,y)
print("The slope and intercept are as per follows {} and {}".format(weight,bias))
```
