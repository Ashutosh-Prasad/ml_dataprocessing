```python
import pandas as pd

df=pd.read_csv("/kaggle/input/carprices/carprices.csv")
df
```

```python
dummies = pd.get_dummies(df["Car Model"])
dummies
```

```python
df_final = pd.concat([df,dummies],axis="columns")
df_final
```

```python
df_final = df_final.drop(["Car Model","Mercedez Benz C class"],axis="columns")
df_final
```

```python
X = df_final.drop(["Sell Price($)"],axis="columns").values
X
```

```python
Y = df["Sell Price($)"].values
Y
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X,Y)
benz = model.predict([[45000,4,0,0]])[0]
bmw = model.predict([[86000,7,0,1]])[0]
accuracy = model.score(X,Y)
```

```python
print("price of benz model is",benz)
print("price of bmw model is",bmw)
print("The accuracy of the model is {}%".format(accuracy*100))
```

```python
x = x[:,1:]
x
```
